Gaussian Discriminant Analysis
----------------

We have two classes from (4). Now, as in (2) we use a GDA model where we assume that the class conditional densities are Gaussian and that the two classes share the same covariance matrix. We estimate the parameters using the maximum likelihood approach.

```python
def learn(training_set):
    
    # Populate list X with data
    data = list(csv.reader(open(training_set)))
    X = []
    for index in range(len(data)):
        X.append(data[index])
    # Convert lists to numpy float arrays
    training_set = np.asarray(X)
    training_set = training_set.astype(np.float)
    
    # Get matrix of training examples. We want to find u0, u1, cov (s0, s1), pclass_0, pclass_1
    total = len(training_set)
    u_0 = 0
    u_1 = 0
    s_0 = 0
    s_1 = 0
    count_0 = 0
    count_1 = 0
    
    # Loop through once to get class counts and sum (for u_0, u_1)
    for row in training_set:
        # Check class label
        if row[-1] == -1:
            u_0 += row[:-1]
            count_0 += 1
        elif row[-1] == 1:
            u_1 += row[:-1]
            count_1 += 1
        else:
            print("Made a mistake with dataset creation!")
     
    # Get averages across the row
    u_0 /= count_0
    u_1 /= count_1
    
    # Posterior probabilities (follow Bernoulli(pi), where pi = N_0/N)
    pclass_0 = count_0 / total
    pclass_1 = 1 - pclass_0
    
    # Loop through again to get (x(n) - u)(x(n) - u) ^T (for s_0, s_1)
    for row in training_set:
        # Check class label
        if row[-1] == -1:
            x_0 = row[:-1] - u_0
            x_0 = x_0[:, np.newaxis]
            s_0 += np.dot(x_0, x_0.T)
        
        elif row[-1] == 1:
            x_1 = row[:-1] - u_1
            x_1 = x_1[:, np.newaxis]
            s_1 += np.dot(x_1, x_1.T)
        else:
            print("Made a mistake with dataset creation!")
        
    # Get s_0, s_1
    s_0 /= count_0
    s_1 /= count_1

    # Now we can find the covariance matrix
    cov = pclass_0 * s_0 + pclass_1 * s_1

    # Finally, get weights w0, w1
    cov_inverse = np.linalg.inv(cov)
    w_1 = np.dot(cov_inverse, (u_0 - u_1))
    w_0 = -1/2 * np.dot(u_0.T, np.dot(cov_inverse, u_0)) + 1/2 * np.dot(u_1.T, np.dot(cov_inverse, u_1)) \
    + np.log(pclass_0/pclass_1)
    print("%%%%%%%%  Weights   %%%%%%%\n")
    print("w_1 : \n", w_1)
    print("w_0 : \n", w_0)
    
    weight_dir = os.path.join(os.getcwd(), 'Assignment2_260601793_5_1_coefficients.txt')
    text_coefs = np.append(w_0, w_1)
    np.savetxt(weight_dir, text_coefs, delimiter=",", header='w_0 (the bias term) first, then w_1 (the feature weights)')
        
    return w_1, w_0
   
```

```python
def get_pclass_0(w_1, w_0, x):
    # Returns probability of class 0 (negative class) given x
    a = np.dot(w_1, x) + w_0   
    return sigmoid(a) 

def prob_class_1(x):
    # Returns probability of class 1 (positive class) given x
    return 1 - pclass_0(x)

def sigmoid(x):
    # Generic sigmoid function (apply as sigmoid(w^T x + w0))
    sig = 1/ (1 + np.exp(-x))
    return sig
```

```python
def evaluate(test_set, w_1, w_0):
    # Given the coeficients we learned, use test set to evaluate the accuracy, precision,
    # recall, and F-measure performance of the model.

    # Populate list X with data
    data = list(csv.reader(open(test_set)))
    X = []
    for index in range(len(data)):
        X.append(data[index])
    
    # Convert lists to numpy float arrays
    test_set = np.asarray(X)
    test_set = test_set.astype(np.float)
    
    tp = 0
    tn = 0
    fp = 0
    fn = 0
    total = 0

    # Go through test set examples. Check label. Get class probability and if it is less than
    # 0.5, choose the negative class 
    for row in test_set:
        target_class = row[-1]
        pclass_0 = get_pclass_0(w_1, w_0, row[:-1])
        total +=1
        if pclass_0 < 0.5:
            if target_class == 1:
                tp += 1
            else:
                fp += 1
        else :
            if target_class == -1:
                tn += 1
            else:
                fn += 1
    
    acc = (tn + tp)/ total
    prec = tp / (tp + fp) 
    recall = tp / (tp + fn)
    fmeasure = 2 * prec * recall / (prec + recall)

    print("\n%%%%%%%%%% GDA Performance Metrics %%%%%%%%%%%%%\n")
    print("Accuracy: {}".format(acc))
    print("Precision: {}".format(prec))
    print("Recall: {}".format(recall))
    print("F-Measure: {}".format(fmeasure))

```

```python
training_set = 'DS2_train_set.csv'
test_set = 'DS2_test_set.csv'
w_1, w_0 = learn(training_set)
evaluate(test_set, w_1, w_0)
```

 K-Nearest Neighbours
---------------------------------------
We have two classes from (4). Now, as in (3), we use a k-NN model where we classify a sample is the test set based on its k-nearest neighbours in the training set.

```python
def init(training_set, val_set, test_set):
    # Init function that, given the .csv files, will return data contained in the .csv file
    # in the form of a numpy array (one for training set, one for test set).
    
    train_data = list(csv.reader(open(training_set)))
    test_data = list(csv.reader(open(test_set)))
    val_data = list(csv.reader(open(val_set)))
    # Populate lists train_X and test_X with data
    train_X = []
    test_X = []
    val_X = []
    for index in range(len(train_data)):
        train_X.append(train_data[index])
    for index in range(len(test_data)):
        test_X.append(test_data[index])
    for index in range(len(val_data)):
        val_X.append(val_data[index])

    # Convert lists to numpy float arrays
    training_set = np.asarray(train_X)
    training_set = training_set.astype(np.float)

    test_set = np.asarray(test_X)
    test_set = test_set.astype(np.float)
    
    val_set = np.asarray(val_X)
    val_set = val_set.astype(np.float)
    
    return training_set, val_set, test_set
```

```python
def kn_neighbours(x, k, train_data):
        # For each example from the test set x, we'll look at the k-nearest neighbours,
        # and then take the average of the k-neighbours' class labels. This is a weighted average
        # (the weight for a class label is inversely proportional to its distance from x). The average
        # class of x's k-nearest neighbours is then returned (-1 if average < 0, 1 if average > 0)
        
        k_nearest = []
        for (idx, ex) in enumerate(train_data):
            ex = ex[:-1]
            dif = x - ex
            distance = np.dot(dif, dif)
            if len(k_nearest) < k:
                k_nearest.append([distance, idx])
            else:
                # Find max distance in list, and kick out; replace with new idx, distance
                k_nearest = sorted(k_nearest, reverse=True)
                for list_idx, (old_distance, old_idx) in enumerate(k_nearest):
                    if old_distance > distance:
                        del(k_nearest[0])
                        k_nearest.append([distance, idx])
                        break
       
        # Now we have a list of the k nearest neighbours; their distance from x and the
        # index in train_data they correspond to.
            
        # Take the mean of the k-nearest complete examples
        total_inv_distance = np.sum(1 / np.asarray(k_nearest)[:,0], axis=0)
        average = 0
        for (distance, idx) in k_nearest:
            weight = ((1 /distance) / total_inv_distance)
            average += train_data[idx, -1]
        average /= k
        
        # Now, average value should be in range (-1, 1)
        if average < 0:
            return -1
        else:
            return 1
```

```python
def optimize(train_data, val_data):
    # Classify the test examples using a k-nearest neighbour approach. The unweighted mean
    # of the labels of the K-NN is taken and the label (-1, 1) that the average is closer to 
    # is assigned to the test set example.
    # We iterate through different values of k and compute the accuracy, precision, recall, and F-
    # measure for each k-value.
    
    print("\n%%%%%%%%%% Searching for Optimum k %%%%%%%%%%%%%\n")
    max_f = 0
    max_k = 0
    
    f_measure_list = []
    k_list = []
    for k in range(1,21):
        tp = 0
        tn = 0
        fp = 0
        fn = 0
        count = 0
        for row in val_data:
            # Look at each test sample and compare to its k-NN in training dataset
            count += 1
            x = row[:-1]
            y = row[-1]
            y_hat = kn_neighbours(x, k, train_data)
            
            if y_hat == -1 and y == -1:
                tn += 1
            elif y_hat == -1 and y == 1:
                fn += 1
            if y_hat == 1 and y == 1:
                tp += 1
            elif y_hat == 1 and y == -1:
                fp += 1
   
        acc = (tn + tp) / count
        prec = tp / (tp + fp) 
        recall = tp / (tp + fn)
        fmeasure = (2 * prec * recall) / (prec + recall)
        if fmeasure > max_f and k % 2 != 0:
            max_f = fmeasure
            max_k = k
        
        f_measure_list.append(fmeasure)
        k_list.append(k)
    
    plt.plot(k_list, f_measure_list)
    plt.xlabel('k values')
    plt.ylabel('F-measure')
    plt.show()    
    print("Optimum F-Measure {} found at k = {}".format(max_f, max_k))
    return max_k
```

```python
def evaluate(train_data, test_data, k):
    # Classify the test examples using a k-nearest neighbour approach. The unweighted mean
    # of the labels of the K-NN is taken and the label (-1, 1) that the average is closer to 
    # is assigned to the test set example.
    # We iterate through different values of k and compute the accuracy, precision, recall, and F-
    # measure for each k-value.
    
    print("\n%%%%%%%%%% Evaluating (with k = {}) %%%%%%%%%%%%%\n".format(k))
    
    tp = 0
    tn = 0
    fp = 0
    fn = 0
    count = 0
    for row in test_data:
        # Look at each test sample and compare to its k-NN in training dataset
        count += 1
        x = row[:-1]
        y = row[-1]
        y_hat = kn_neighbours(x, k, train_data)
            
        if y_hat == -1 and y == -1:
            tn += 1
        elif y_hat == -1 and y == 1:
            fn += 1
        if y_hat == 1 and y == 1:
            tp += 1
        elif y_hat == 1 and y == -1:
            fp += 1
   
    acc = (tn + tp) / count
    prec = tp / (tp + fp) 
    recall = tp / (tp + fn)
    fmeasure = (2 * prec * recall) / (prec + recall)
        
    print("K : {}".format(k))
    print("Accuracy: {}".format(acc))
    print("Precision: {}".format(prec))
    print("Recall: {}".format(recall))
    print("F-Measure: {}\n".format(fmeasure))
 
    return None
```

```python
# Initialize data
training_set = 'DS2_train_set.csv'
val_set = 'DS2_val_set.csv'
test_set = 'DS2_test_set.csv'
training_set, val_set, test_set = init(training_set, val_set, test_set)
        
# Now, find the best k:
k = optimize(training_set, val_set)
evaluate(training_set, test_set, k)
```
