```python
import numpy as np
import csv
import os
import matplotlib.pyplot as plt
```

This experiment uses Dataset-2, which consists of train, validation, and test files. The input is a real valued scalar and the output is also a real valued scalar.

1. Fit a linear regression model to this dataset by using stochastic gradient descent (one example at a time).
    a. Use the step size of 1e-6. Compute the MSE on validation set for every epoch.
    b. Plot the learning curve i.e. training and validation MSE for every epoch.
    
2. Try different step sizes and choose the best step size by using validation data.
    a. Report in a table the validation performance with different step-sizes. 
    b. Report the test MSE of the chosen model.

3. Visualize the fit for every epoch and report 5 visualizations chosen at random to illustrate how the regression fit evolves during the training process.


```python
def init(dataset):
    # Initializes the desired training set and returns data in a usable form
    # Input: 
    # dataset - the directory where the .csv file is 
    # Outputs: 
    # X - an n x 1 vector (the independent variable in the dataset)
    # Y - an n x 1 vector (the dependent variable in the dataset that 
    # corresponds to X)
    
    # Read in data with csv reader
    data = list(csv.reader(open(dataset)))
    
    # Populate lists X and Y with data
    X = []
    Y = []
    for index in range(len(data)):
        X.append(data[index][0])
        Y.append(data[index][1])
    
    # Convert lists to numpy float arrays
    X = np.asarray(X)
    Y = np.asarray(Y)
    X = X.astype(np.float)
    Y = Y.astype(np.float)
    return X, Y
```

```python
def mean_squared_error(X, Y, W):
    # Implements the mean squared error function, i.e.:
    # E = 1/N sum from i = 0 to N [X*W^T - Y]^2
    # Inputs:
    # X - an n x 1 vector (the independent variable)
    # W - a degree x 1 vector (the weights that map X to Y_hat)
    # Y - an n x 1 vector (the dependent variable in the dataset that 
    # corresponds to X)
    # degree - an integer (the degree of the polynomial we're trying to 
    # fit the data with)
    # Outputs:
    # loss - a float loss computed from the MSE equation (above)
    
    # Map X to Y_hat with the current weight vector
    Y_hat = polynomial(X, W)
    inner = np.power((Y_hat - Y), 2)
    loss = np.sum(inner) / len(X)
    return loss
```

```python
def polynomial(X, W):
    # Returns a polynomial for X values for coefficients W. It is assumed
    # W values are in order of increasing degree
    # Inputs:
    # X - an n x 1 vector (the independent variable)
    # W - a degree x 1 vector (the weights that map X to Y_hat)
    # Outputs:
    # Y_hat - an n x 1 vector Y_hat[i] = X[i] * W[0] + ... + 
    # X[i]**degree * W[degree]
    
    degree = len(W)
    Y_hat = 0
    for n in range(degree):
        Y_hat += W[n] * X**n
    return Y_hat    
```

```python
def stochastic_grad_descent(X, Y, X_val, Y_val, alpha=1e-2, epochs=20, 
                     regularization=False, plot_loss = False, evolution_plot=False, 
                     degree=1, L=0.1):
    
    # Performs stochastic gradient descent on inputs (X,Y). For this example
    # we use degree = 1 (for linear regression). Each epoch iterates through
    # all examples (X(i), Y(i)) and updates the weights as follows:
    # w0 = w0 - alpha * (X(i) * w1 + w0 - Y(i))
    # w1 = w1 - alpha * X(i) * (X(i) * w1 + w0 - Y(i))
    # where alpha is the learning rate.
    # Inputs:
    # X - an n x 1 vector (independent variable) belonging to training set
    # Y - an n x 1 vector (dependent variable) belonging to training set
    # X_val - an m x 1 vector (independent variable) belonging to validation set
    # Y_val - an m x 1 vector (dependent variable) belonging to validation set
    # alpha - the learning rate or step size (see equations above)
    # epochs - number of epochs to perform
    # reguarization - if true, incorporate L2 regularization in the sgd updates
    # plot_loss - if true, plots MSE as a function of epoch number
    # evolution_plot - if true, plots the regression fit at specified intervals
    # degree - the degree of the polynomial we are using to fit the data
    # L - the lambda coefficient for regularization
    # Outputs:
    # W - a degree x 1 vector of the learned weights w[0], w[1], ... w[degree]

    
    train_loss_list = []
    val_loss_list = []
    W = np.array([1.,1.])
    for epoch in range(epochs + 1):
        # Update the parameters W here following stochastic gradient descent
        # algorithm presented in lecture, w = w - alpha * loss(w)
        shuffled_examples = [example for example in range(len(X))]
        np.random.shuffle(shuffled_examples)
        for i in shuffled_examples:
            error = polynomial(X[i], W) - Y[i]
            for n in range(degree + 1):
                if not regularization:
                    W[n] = W[n] - ((alpha) * (X[i] ** n) * error)
                else:
                    W[n] = W[n] - ((alpha) * ((X[i] ** n) * error + (L * W[n])))

        # Update the loss here
        train_loss = mean_squared_error(X, Y, W)
        val_loss = mean_squared_error(X_val, Y_val, W)
        
        # Plot regression line as fit evolves during training
        for sample in [0, 1, 2, 6, 12]:
            if sample == epoch and evolution_plot==True:
                # Plot regression line over training data
                x = np.linspace(0,1.5)
                plt.plot(x, polynomial(x, W))
                plt.plot(X_train, Y_train, 'rx')
                plt.xlabel('Input X')
                plt.ylabel('Output Y')
                plt.ylim([3.5, 10.5])
                plt.title('Training Data and SGD Line Fit after Epoch {}'.format(epoch))
                plt.savefig('sgd_fit_train_epoch{}.png'.format(epoch))
                plt.show()

        train_loss_list.append(train_loss)
        val_loss_list.append(val_loss)
        if epoch == epochs:
        # Plot learning curve for training and validation
                epoch_list = [epoch for epoch in range(epochs + 1)]
                plt.plot(epoch_list, train_loss_list, 'r', label='training')
                plt.plot(epoch_list, val_loss_list, 'b', label='validation')
                plt.xlabel('Epoch')
                plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
                plt.ylabel('Mean Squared Error')
                plt.ylim([0, 0.6])
                plt.title('Learning Curve for Training Data and Validation Data')
                plt.legend(loc='upper right')
                plt.savefig('sgd_fit_train_valid.png')
                plt.show() 
    return W
```

```python
#Training data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_2_train.csv')
X_train, Y_train = init(data_path)

#Validation data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_2_valid.csv')
X_val, Y_val = init(data_path)

#Test data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_2_test.csv')
X_test, Y_test = init(data_path)


W = stochastic_grad_descent(X_train, Y_train, X_val, Y_val, 
                           plot_loss=True, evolution_plot=True)
```

```python
# Plot final regression line over training data
x = np.linspace(0,1.5)
plt.plot(x, polynomial(x, W))
plt.plot(X_train, Y_train, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([3.5, 10.5])
plt.title('Training Data and SGD Line Fit to Minimize MSE')
plt.savefig('sgd_fit_train.png')
plt.show()

# Teraining MSE of our model:
train_loss = mean_squared_error(X_train, Y_train, W)
print("Train loss is : ", train_loss)

# Validation MSE of our model:
val_loss = mean_squared_error(X_val, Y_val, W)
print("Validation loss is : ", val_loss)

# Test MSE of our model:
test_loss = mean_squared_error(X_test, Y_test, W)
print("Test loss is : ", test_loss)
```
