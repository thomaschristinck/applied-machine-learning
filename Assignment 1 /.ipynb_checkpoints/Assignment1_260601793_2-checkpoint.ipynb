```python
import numpy as np
import csv
import os
import matplotlib.pyplot as plt
```

Dataset-1 consists of train, validation, and test files. The input is a real valued scalar and the output is also a real valued scalar. The dataset is generated from an n-degree polynomial and a small Gaussian noise is added to the target.

1. Fit a 20-degree polynomial to the data.
    1. Report the training and validation MSE (Mean-Square Error). Do not use any regularization.
    2. Visualize the fit.
    3. Is the model overfitting or underfitting ? Why?


2. Now add L2 regularization to your model. Vary the value of λ from 0 to 1.
    1. For different values of λ, plot the training MSE and the validation MSE
    2. Pick the best value of λ and report the test performance for the corresponding model.
    3. Also visualize the fit for the chosen model.

```python
def init(training_set):
    # Initializes the desired training set and returns data in a usable form
    # Input: 
    # Training set - the directory where the .csv file is 
    # Outputs: 
    # X - an n x 1 vector (the independent variable in the dataset)
    # Y - an n x 1 vector (the dependent variable in the dataset that 
    # corresponds to X)
    
    # Read in data with csv reader
    data = list(csv.reader(open(training_set)))
    
    # Populate lists X and Y with data
    X = []
    Y = []
    for index in range(len(data)):
        X.append(data[index][0])
        Y.append(data[index][1])
    
    # Convert lists to numpy float arrays
    X = np.asarray(X)
    Y = np.asarray(Y)
    X = X.astype(np.float)
    Y = Y.astype(np.float)
    return X, Y
```

```python
def mean_squared_error(X, Y, W):
    # Implements the mean squared error function, i.e.:
    # E = 1/N sum from i = 0 to N [X*W^T - Y]^2
    # Inputs:
    # X - an n x 1 vector (the independent variable)
    # W - a degree x 1 vector (the weights that map X to Y_hat)
    # Y - an n x 1 vector (the dependent variable in the dataset that 
    # corresponds to X)
    # degree - an integer (the degree of the polynomial we're trying to 
    # fit the data with)
    # Outputs:
    # loss - a float loss computed from the MSE equation (above)
    
    # Map X to Y_hat with the current weight vector
    Y_hat = polynomial(X, W)
    inner = np.power((Y_hat - Y), 2)
    loss = np.sum(inner) / len(X)
    return loss
```

```python
def polynomial(X, W):
    # Returns a polynomial for X values for coefficients W. It is assumed
    # W values are in order of increasing degree
    # Inputs:
    # X - an n x 1 vector (the independent variable)
    # W - a degree x 1 vector (the weights that map X to Y_hat)
    # Outputs:
    # Y_hat - an n x 1 vector Y_hat[i] = X[i] * W[0] + ... + 
    # X[i]**degree * W[degree]
    
    degree = len(W)
    Y_hat = 0
    for n in range(degree):
        Y_hat += W[n] * X**n
    return Y_hat          
```

```python
def poly_fit(X_in, Y, degree, regularization=0):
    # Minimizes the mean squared error of some function Y = X * W^T 
    # This was derived in class as W = (X^T X)^-1 * X^T * Y
    # First we need to make a matrix X where the rows consist of each X 
    # and the columns consist of each power of X from 0 to "degree"
    
    X = np.zeros((len(X_in), degree + 1))
    for index, item in enumerate(X_in):
        for n in range(degree + 1):
            X[index][n] = item ** n
    
    temp1 = np.matmul(X.transpose(), X) 
    temp1 += (np.identity(X.shape[1]) * regularization) 
    temp2 = np.matmul(X.transpose(),Y)
    W = np.matmul(np.linalg.pinv(temp1), temp2)
   
    return W
    
```

```python
##################################################################
#  POLYNOMIAL REGRESSION (without regularization)
##################################################################

#Training data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_1_train.csv')
X, Y = init(data_path)

#Validation data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_1_valid.csv')
X_val, Y_val = init(data_path)

#Test data
data_path = os.path.join(os.getcwd(), 'Datasets/Dataset_1_test.csv')
X_test, Y_test = init(data_path)

W = poly_fit(X, Y, 20)

# Plot regression curve over training data
x = np.linspace(-1,1)
plt.plot(x, polynomial(x, W))
plt.plot(X, Y, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([-15, 35])
plt.title('Training Data and Polynomial Fit to Minimize MSE')
plt.savefig('polynomial_fit_train.png')
plt.show()

train_loss = mean_squared_error(X, Y, W)
print("Training loss is ", train_loss)

# Validation data
plt.plot(x, polynomial(x, W))
plt.plot(X_val, Y_val, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([-15, 35])
plt.title('Validation Data and Polynomial Fit to Minimize MSE')
plt.savefig('polynomial_fit_valid.png')
plt.show()
val_loss = mean_squared_error(X_val, Y_val, W)
print("Validation loss is ", val_loss)


# Test data
plt.plot(x, polynomial(x, W))
plt.plot(X_test, Y_test, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([-15, 35])
plt.title('Test Data and Polynomial Fit to Minimize MSE')
plt.savefig('polynomial_fit_test.png')
plt.show()

test_loss = mean_squared_error(X_test, Y_test, W)
print("Test loss is ", test_loss)
```

```python
##################################################################
#  REGRESSION (with regularization)
##################################################################

# Explore varying regularization coefficients and get optimum
lambdas = np.linspace(0,1, 100)
weight_list = []
train_loss_list = []
val_loss_list = []
for L in lambdas:
    W = poly_fit(X, Y, degree=20, regularization = L)
    weight_list.append(W)
    train_loss = mean_squared_error(X, Y, W)
    val_loss = mean_squared_error(X_val, Y_val, W)
    train_loss_list.append(train_loss)
    val_loss_list.append(val_loss)

# Plot optimum fit on training data
min_index = np.argmin(val_loss_list)
lambda_min = lambdas[min_index]
w_min = weight_list[min_index]
plt.plot(x, polynomial(x, w_min))
plt.plot(X, Y, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.xlim([-1.05, 1.05])
plt.ylim([-15, 35])
plt.title('Polynomial Fit to Minimize MSE with Training Data and Lambda {0:.3f}'.format(lambda_min))
plt.savefig('polynomial_fit_train.png')
plt.show()

train_loss = mean_squared_error(X, Y, w_min)
print("Training loss is ", train_loss)

#Plot optimum fit on validation data
plt.plot(x, polynomial(x, w_min))
plt.plot(X_val, Y_val, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([-15, 35])
plt.title('Polynomial Fit to Minimize MSE with Validation Data and Lambda {0:.3f}'.format(lambda_min))
plt.savefig('polynomial_fit_valid.png')
plt.show()

val_loss = mean_squared_error(X_val, Y_val, w_min)
print("Validation loss is ", val_loss)

#Plot optimum fit on test data
plt.plot(x, polynomial(x, w_min))
plt.plot(X_test, Y_test, 'ro')
plt.xlabel('Input X')
plt.ylabel('Output Y')
plt.ylim([-15, 35])
plt.title('Polynomial Fit to Minimize MSE with Test Data and Lambda {0:.3f}'.format(lambda_min))
plt.savefig('polynomial_fit_test.png')
plt.show()

test_loss = mean_squared_error(X_test, Y_test, w_min)
print("Test loss is ", test_loss)

# Plot regularization coeffient vs MSE
plt.plot(lambdas, train_loss_list, 'r', label='training')
plt.plot(lambdas, val_loss_list, 'b', label='validation')
plt.xlabel('Regularization Coefficient')
plt.ylabel('Mean Squared Error')
plt.ylim([7, 11.5])
plt.title('Training and Validation Error as Regularization Coefficient Varies')
plt.legend(loc='lower right')
plt.savefig('polynomial_fit_train&valid_regularization.png')
plt.show()

print("Optimal lambda is {0:.3f}".format(lambda_min))
```
